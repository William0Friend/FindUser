## 1. Pie Plot

**Syntax**: ax.pie(x, labels, Colors, Autopct)

|Argument|Description|
|:--|:-------------------------------|
|x	|array-like. The wedge sizes.|
|labels	|list. A sequence of strings providing the labels for each wedge.|
|Colors	|A sequence of matplotlibcolorargs through which the pie chart will cycle. |
||If None, will use the colors in the currently active cycle.|
|Autopct |string, used to label the wedges with their numeric value. The label will be placed |
||inside the wedge. The format string will be fmt%pct.|

```python
fig, ax = plt.subplots(figsize=(5,5))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
pie = plt.pie(students, labels = langs,autopct='%1.2f%%')
```

```python
fracs = [30, 15, 45, 10]
colors = ['b', 'g', 'r', 'w']

fig, ax = plt.subplots(figsize=(5, 5))  # make the plot square
pie = ax.pie(fracs, colors=colors, explode=(0, 0, 0.15, 0), shadow=True,
             labels=['A', 'B', 'C', 'D'])
```

## 2. Errorbar Plots

```python
x = np.linspace(0, 10, 30)
dy = 0.2
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots(figsize=(6,3))
plt.errorbar(x, y, dy, fmt='.b')
```

## 3. Bar Plot

### 3.1. Vertical Bar

**Syntax**: ax.bar(x, height, width, bottom, align)

|Argument|Description|
|:--|:-------------------------------|
|x	|sequence of scalars representing the x coordinates of the bars. align controls if x is the bar| |center |(default) or left edge.|
|height	|scalar or sequence of scalars representing the height(s) of the bars.|
|width	|scalar or array-like, optional. the width(s) of the bars default 0.8|
|bottom|	scalar or array-like, optional. the y coordinate(s) of the bars default None.|
|align|	{‘center’, ‘edge’}, optional, default ‘center’|

```python
fig, ax = plt.subplots(figsize=(6,3))

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students, width=0.2)
```

```python
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)

fig, ax = plt.subplots(figsize=(8,4))

ax.bar(X + 0.00, data[0], color = 'b', width = 0.25) # four blue bars
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25) # four green bars
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25) # four red bars
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups
width = 0.35

fig, ax = plt.subplots(figsize=(8,4))
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width, bottom=menMeans, color='b') # bottom option gives us stacked bar

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
```

### 3.2. Horizontal Bar

```python
Product = ['Computer','Monitor','Laptop','Printer','Tablet']
Quantity = [320,450,300,120,280]

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8,3))

plt.barh(Product,Quantity) # barh means horizontal bar
plt.title('Store Inventory')
plt.ylabel('Product')
plt.xlabel('Quantity')
```

**Note**: You can plot the same bar chart with the help of the Pandas library

```python

# You can plot the same bar chart with the help of the Pandas library
data = {'Quantity': [320,450,300,120,280],
        'Price': [800,250,1200,150,300]
       }
df = pd.DataFrame(data,columns=['Quantity','Price'], index = ['Computer','Monitor','Laptop','Printer','Tablet'])

# dataframe plot with df.plot.???
ax = df.plot.barh(figsize=(5, 3))

ax.set_title('Store Inventory')
ax.set_ylabel('Product')
ax.set_xlabel('Quantity')
plt.show()
```

## 4.  Contour Plot

meshgrid:

```python
plt.style.use('classic')

xlist = np.linspace(-2.0, 2.0, 20)
ylist = np.linspace(-2.0, 2.0, 20)
X, Y = np.meshgrid(xlist, ylist)

# look at mesh grids
fig, ax = plt.subplots(figsize=(6,6))
plt.scatter(X,Y)
plt.axis('equal')
#plt.xlim(-2.5,2.5)

# now compute Z (function(x,y)) for each grid
Z = np.sqrt(X**2 + Y**2)

# build contour plot
fig,ax=plt.subplots(figsize=(6,6))
cp = ax.contourf(X, Y, Z)

# add label/title/colorbar...
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
ax.set_aspect('equal')
```

```python

x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)

z = np.cos(y[:, np.newaxis]) * np.sin(x)

fig, ax = plt.subplots()

# filled contours
im = ax.contourf(x, y, z, 100)

# contour lines
im2 = ax.contour(x, y, z, colors='g')

fig.colorbar(im, ax=ax)
```

## Box Plot

Understanding Boxplots: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51

```python
plt.style.use('seaborn-whitegrid')
np.random.seed(10)
c1 = np.random.normal(100, 10, 20)
c2 = np.random.normal(80, 30, 200)
c3 = np.random.normal(90, 20, 200)
c4 = np.random.normal(70, 25, 200)

#print(c1)

fig, ax = plt.subplots(figsize=(4,3))
bp = ax.boxplot([c1, c2, c3, c4])
plt.show()
```

## Violin Plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# Create a figure instance
fig, ax = plt.subplots(figsize=(4,3))

# Create the boxplot
bp = ax.violinplot(data_to_plot)
```
